```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Covid 19 dataset\557629_1206897_bundle_archive\covid_19_india.csv",parse_dates=["Date"])
df.set_index('Date',inplace=True)
```

```python
df.columns
```

```python
g = df.groupby("State/UnionTerritory")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#pass the list of city here

```python
State1 = ['Haryana','Punjab']
for i, val in enumerate(State1):
    if i == 0:
        State1_plot = g.get_group(val)
        State1_plot = State1_plot.loc['05-20-2020':'05-30-2020']
        State1_plot.reset_index(inplace=True)
    else:
        State2_plot = g.get_group(val)        
        State2_plot = State2_plot.loc['05-20-2020':'05-30-2020']
        State2_plot.reset_index(inplace=True)
```

```python
plt.figure(figsize =(12,10))
xpos = np.arange(len(State1_plot.Date))

plt.xticks(xpos,State1_plot.Date,rotation='vertical')
graph1 = plt.bar(xpos-0.25,State1_plot.Confirmed,width = 0.4,label =  'State1')
graph2 = plt.bar(xpos+0.25,State2_plot.Confirmed,width = 0.4,label =  'State2')

plt.ylabel('Confirmed Cases')
plt.title('Comparison of Cases in two States')
plt.legend()
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(graph1)
autolabel(graph2)
```
